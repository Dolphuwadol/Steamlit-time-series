### เขียนให้อยู๋ในรูป funtion

```python
def get_set50():
    driver = webdriver.Chrome()
    driver.get("https://www.set.or.th/th/market/index/set50/overview")
    return driver
```

```python
driver = get_set50()
```

```python
def get_set50_df(driver):
    data = driver.page_source
    df = pd.read_html(data)[1]
    clean_column = []
    for c in df.columns:
        clean_column.append(c.replace(' (Click to sort Ascending)',''))
    df.columns = clean_column
    df = df.set_index('หลักทรัพย์ ')
    return df
```

```python
df = get_set50_df(driver)
df
```

```python
def get_stock_data(stock, driver, all_stock_dict):  
    stock = stock.split()[0] # clean ข้อมูลพวก xd xr ที่อยู่หน้าชื่อหุ้นออกไป
    url = 'https://www.set.or.th/th/market/product/stock/quote/' + stock + '/financial-statement/company-highlights'
    driver.get(url)
    stock_data = driver.page_source
    financial = pd.read_html(stock_data)[0]
    stat = pd.read_html(stock_data)[1]
    all_stock_dict[stock] = [financial, stat]
    return driver, all_stock_dict
```

```python
all_stock_dict = dict()
for stock in df.index:
    driver, all_stock_dict = get_stock_data(stock, driver, all_stock_dict)
```
